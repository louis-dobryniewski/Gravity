```python
import pandas as pd
import Gravity as grav
```

```python
# Reading in planetary data from the nasa website
table = pd.read_html('https://nssdc.gsfc.nasa.gov/planetary/factsheet/')[0]
pd.DataFrame(table.values.T,columns=table[0]).drop(0)
```

```python
earth = grav.Planet(name='earth', mass=5.972e24,position=(0,0),velocity=(0,-12))
moon = grav.Planet(name='moon', mass=7.34767309e22,position=(0.3633e9,-268.5e6),velocity=(0,1082))
ship = grav.Planet(name='ship', mass=3000,position=(6371e3,0),velocity=(11.6e3,-12))
planets = [earth,moon,ship]

grav.DELTA = 1000

earth_moon_system = grav.System(planets)
earth_moon_system.run(time=100_000)
earth_moon_system.plot(x_lim=(-1e9,0.5e9), y_lim=(-0.15e10,0.5e9))
```

```python
sun =       grav.Planet(name='sun', mass=1.989e30,position=(0,0),velocity=(0,0))
mercury =   grav.Planet(name='mercury', mass=3.285e23,polar_position=(46.0012e9,0),max_speed=58.98e3)
venus =     grav.Planet(name='venus', mass=4.87e24,polar_position=(107.5e9,0),max_speed=35.0e3)
earth =     grav.Planet(name='earth', mass=5.97e24,polar_position=(147.1e9,0),max_speed=29.8e3)
mars =      grav.Planet(name='mars', mass=0.64169e24,polar_position=(206.6509e9,0),max_speed=26.50e3)
pluto =     grav.Planet(name='pluto', mass=0.0130e24, polar_position=(4436.8e9,0), max_speed=6.1e3)
jupyter =   grav.Planet(name='jupyter', mass=1898.13e24,polar_position=(740.595e9,0), max_speed=13.72e3)
saturn =    grav.Planet(name='saturn', mass=568.32e24, polar_position=(1357.554e9,0), max_speed=10.18e3)
uranus =    grav.Planet(name='uranus', mass=86.811e24, polar_position=(2732.696e9,0), max_speed=7.11e3)
neptune =   grav.Planet(name='neptune', mass=102.409e24, polar_position=(4471.050e9,0),max_speed=5.50e3)
planets = [sun, mercury, venus, earth, mars, jupyter, saturn, uranus, pluto]

grav.DELTA = 50000
grav.GRAV = 6.674e-11

solar_system = grav.System(planets)
solar_system.run(time=6_000)
solar_system.plot()
```
